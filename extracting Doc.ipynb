```python
def extract_text_from_docx(docx_path):
    # Load the document
    doc = Document(docx_path)

    # Extract text from each paragraph
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    
    # Join the list into a single string with newline characters
    return '\n'.join(full_text)
```

```python
def extract_tables_from_docx(docx_path):
    # Load the Word document
    doc = Document(docx_path)
    # Initialize an empty list to store all tables
    tables = []
    # Iterate over each table in the document
    for table in doc.tables:
        # Initialize an empty list to store the current table's data
        table_data = []
        # Iterate over each row in the current table
        for row in table.rows:
            # Initialize an empty list to store the current row's data
            row_data = []
            # Iterate over each cell in the current row
            for cell in row.cells:
                # Append the text content of the current cell to the row_data list
                row_data.append(cell.text)
            # Append the current row's data to the table_data list
            table_data.append(row_data)
        # Append the current table's data to the tables list
        tables.append(table_data)
    
    return tables
```

```python
# Path to the Word document
docx_path = "C:\\Users\\Local User\\Pictures\\assignment seu\DBL final update 2021100000002-sheikh md. salman farsi.docx"

# Extract and print the tables
tables = extract_tables_from_docx(docx_path)

# Iterate over each table to print its content
for i, table in enumerate(tables):
    print(f"Table {i + 1}")
    # Iterate over each row in the current table
    for row in table:
        print(row)
```

```python
# Import the Document class from the python-docx library to work with Word documents
from docx import Document

# Import functions and constants to work with XML namespaces and XML parsing
from docx.oxml.ns import qn
from docx.oxml import parse_xml

# Import the os library for interacting with the operating system, such as file and directory manipulation
import os

def extract_images_from_docx(docx_path, output_dir):
    # Load the Word document using the Document class from python-docx
    doc = Document(docx_path)
    # Ensure the output directory exists, create it if it doesn't
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    # Initialize a counter for naming the images sequentially
    image_counter = 1
    # Iterate over all relationships (rels) in the document
    for rel in doc.part.rels.values():
        # Check if the relationship target reference contains "image", indicating it is an image
        if "image" in rel.target_ref:
            # Get the binary data of the image
            image_data = rel.target_part.blob
            # Determine the image format based on the file extension (default to jpg if unknown)
            image_format = rel.target_ref.split('.')[-1]
            if image_format not in ['png', 'jpg', 'jpeg', 'gif', 'bmp']:
                image_format = 'jpg'
            # Create a filename for the image using the counter
            image_filename = f'image_{image_counter}.{image_format}'
            image_counter += 1
            # Construct the full path for the image file
            image_path = os.path.join(output_dir, image_filename)
            # Save the image to the output directory
            with open(image_path, 'wb') as img_file:
                img_file.write(image_data)
            # Print the path of the saved image for verification
            print(f'Saved: {image_path}')

# Path to the Word document
docx_path = "C:\\Users\\Local User\\Pictures\\assignment seu\DBL final update 2021100000002-sheikh md. salman farsi.docx"

# Directory to save the extracted images
output_dir = "C:\\Users\\Local User\\Desktop\\python_2308\\images"

# Call the function to extract and save the images
extract_images_from_docx(docx_path, output_dir)
```
